Now, let's write a function that returns car data given a link

```python
def get_car_info(car_link):
    car_dict = {}
    try:
        response = requests.get(car_link)
        if not response.status_code == 200:
            return car_dict
        car_soup = BeautifulSoup(response.content,'html.parser')
        try:
            car_dict['Model']=car_soup.find('div',class_="d-flex justify-content-between").find_all('h2')[0].get_text()
        except:
            car_dict['Model']=None
        try:
            car_dict['Price']=car_soup.find('div',class_="d-flex justify-content-between").find_all('h2')[1].get_text()
        except:
            car_dict['Price']=None
        table = car_soup.find('table',class_="table table-sm mb-4") 
        for row in table.find_all('tr'):  
            cells = row.find_all('td')
            if len(cells) == 2:
                key = cells[0].text.strip()
                value = cells[1].text.strip()
                car_dict[key] = value
        try:
            car_dict['Pic_num']=len(car_soup.find('div',class_='col-12 d-flex mt-3 justify-content-center flex-wrap').find_all('div',class_="justify-content-center px-1"))
        except:
            car_dict['Pic_num']=None
        try:
            car_dict['Description']=car_soup.find('p',class_='text-word-break').get_text().strip()
        except:
            car_dict['Description']=None
        try:
            date_string = car_soup.find_all('div',class_='px-3')[0].get_text()
            date_pattern = r'\b\d{2}/\d{2}/\d{4}\b'
            match = re.search(date_pattern, date_string)
            if match:
                car_dict['Cre_date']= match.group()
            else:
                car_dict['Cre_date']= None
        except:
            car_dict['Cre_date']= None
        try:
            date_string2 = car_soup.find_all('div',class_='px-3')[1].get_text()
            match2 = re.search(date_pattern, date_string2)
            if match2:
                car_dict['Repub_date']= match2.group()
            else:
                car_dict['Repub_date']= None
        except:
            car_dict['Repub_date']= None
        return car_dict
    except:
        return car_dict
```

Reorder columns

```python
Merc_df=Merc_df[['Manufactor','Year','Model','Hand','Gear','capacity_Engine','Engine_type','Prev_ownership','Curr_ownership','Area','City','Price','Pic_num','Cre_date','Repub_date','Description','Color','Km','Test']]
Merc_df
```

```python
#getting off from commas and shekel sign in Price column
Merc_df['Price'] = Merc_df['Price'].str.replace(',', '')
Merc_df['Price'] = Merc_df['Price'].str.replace('₪', '')
```

```python
#getting off from commas sign in capacity_Engine column
Merc_df['capacity_Engine'] = Merc_df['capacity_Engine'].str.replace(',', '')
```

```python
#getting off from commas sign in Km column
Merc_df['Km'] = Merc_df['Km'].str.replace(',', '')
```

```python
#In order to transofrm type with nulls with must use "Int32" type otherwise we will have errors

Merc_df['Manufactor'] = Merc_df['Manufactor'].astype(str)
Merc_df['Year'] = pd.to_numeric(Merc_df['Year'], errors='coerce').astype('Int32')
Merc_df['Model'] = Merc_df['Model'].astype(str)
Merc_df['Hand'] = pd.to_numeric(Merc_df['Hand'], errors='coerce').astype('Int32')
Merc_df['Gear'] = Merc_df['Gear'].astype('category')
Merc_df['capacity_Engine'] = pd.to_numeric(Merc_df['capacity_Engine'], errors='coerce').astype('Int32')
Merc_df['Engine_type'] = Merc_df['Engine_type'].astype('category')
Merc_df['Prev_ownership'] = Merc_df['Prev_ownership'].astype('category')
Merc_df['Curr_ownership'] = Merc_df['Curr_ownership'].astype('category')
Merc_df['Area'] = Merc_df['Area'].astype(str)
Merc_df['City'] = Merc_df['City'].astype(str)
Merc_df['Price'] = Merc_df['Price'].astype(float)
Merc_df['Pic_num'] = pd.to_numeric(Merc_df['Pic_num'], errors='coerce').astype('Int32')
Merc_df['Cre_date'] = pd.to_datetime(Merc_df['Cre_date'], format='%d/%m/%Y') #must use this specific date format
Merc_df['Repub_date'] = pd.to_datetime(Merc_df['Repub_date'], format='%d/%m/%Y') #must use this specific date format
Merc_df['Description'] = Merc_df['Description'].astype(str)
Merc_df['Color'] = Merc_df['Color'].astype(str)
Merc_df['Km'] = pd.to_numeric(Merc_df['Km'], errors='coerce').astype('Int32')
Merc_df['Test'] = pd.to_datetime(Merc_df['Test'], format='%m/%Y') + pd.offsets.MonthEnd(0)#must use this specific date format
Merc_df['Test'] = (Merc_df['Test'] - pd.Timestamp.today()).dt.days.astype("Int32")
```
