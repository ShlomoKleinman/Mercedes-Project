### Make list of all cars with links

page 1

```python
car_list = list()      
cars = soup.find_all('div',class_="card-body p-md-3")
for car in cars:
    link = "https://www.ad.co.il" + car.find('a').get('href')
    name = car.find('a').get_text()
    car_list.append((name,link)) 


```

page 2

```python
url = "https://www.ad.co.il/car?keyword=%D7%9E%D7%A8%D7%A6%D7%93%D7%A1&pageindex=2"
page2 = requests.get(url)
soup2 = BeautifulSoup(page2.content,'html.parser')
cars2 = soup2.find_all('div',class_="card-body p-md-3")
for car in cars2:
    link = "https://www.ad.co.il" + car.find('a').get('href')
    name = car.find('a').get_text()
    car_list.append((name,link))
```

page 3

```python
url = "https://www.ad.co.il/car?keyword=%D7%9E%D7%A8%D7%A6%D7%93%D7%A1&pageindex=3"
page3 = requests.get(url)
soup3 = BeautifulSoup(page3.content,'html.parser')
cars3 = soup3.find_all('div',class_="card-body p-md-3")
for car in cars3:
    link = "https://www.ad.co.il" + car.find('a').get('href')
    name = car.find('a').get_text()
    car_list.append((name,link))
car_list
```

```python
sorted(car_list)
```

Filter non Mercedes

```python
filtered_list = [car for car in car_list if 'מרצ' in car[0]]
sorted(filtered_list)
```

Remove duplicates

```python
seen = set()
unique_list = []
for car in filtered_list:
    if car not in seen:
        unique_list.append(car)
        seen.add(car)

sorted(unique_list)
```

#### Now that we have our list, let's find the data in every car

```python
url = "https://www.ad.co.il/ad/15719227"
car_page = requests.get(url)
car_soup = BeautifulSoup(car_page.content,'html.parser')
print(car_soup.prettify())
```

```python
print(car_soup.find('table',class_="table table-sm mb-4").prettify())
```

```python
table = car_soup.find('table',class_="table table-sm mb-4") 
# Extract rows
rows = []
for tr in table.find_all('tr'):  
    cells = tr.find_all('td')
    row = [cell.text.strip() for cell in cells]
    rows.append(row)
rows



```
